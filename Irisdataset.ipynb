```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
```

```python
iris = load_iris()
print(iris)
```

```python
# Create a dataframe from the iris dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Add 'Target' column as 'Flower' attribute
df['Flower'] = pd.Series(iris.target).replace({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})
```

```python
# Display first 5 entries of the dataframe
print(df.head())
```

```python
# Display the frequency (count) of each flower
print(df['Flower'].value_counts())
```

```python
# Display scatter plot between Petal Length and Petal Width
plt.scatter(df['petal length (cm)'], df['petal width (cm)'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Display Histogram for Petal Length using the library matplotlib
plt.hist(df['petal length (cm)'])
plt.xlabel('petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display Histogram for Petal Width using the library seaborn
sns.histplot(df, x='petal width (cm)')
plt.xlabel('petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display Boxplot for the attributes: Sepal Length, Sepal Width, Petal Length, and Petal Width
sns.boxplot(data=df)
plt.show()
```

```python
# Create and display the correlation matrix with different colors for the attributes: Sepal Length, Sepal Width, Petal Length, and Petal Width
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Create Heatmap for the correlation using the library seaborn
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.show()
```

```python
# Create a simple Linear Regression Model to predict the Petal Width using the Petal Length
X = df[['petal length (cm)']]
y = df['petal width (cm)']
model = LinearRegression().fit(X, y)
```

```python
# Display scatter plot showing the regression model
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('petal Length (cm)')
plt.ylabel('petal Width (cm)')
plt.show()
```

```python
# Display graph for the bell curve (normal distribution)
sns.displot(df['petal length (cm)'], kde=True)
plt.xlabel('petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
